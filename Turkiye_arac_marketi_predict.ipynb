```python
#Verileri uygun değişken ismine atıyoruz
names = data.columns.tolist()
names[names.index("Yakıt Turu")] = "Yakıt_Turu"
names[names.index("İlan Tarihi")] = "İlan_Tarihi"
names[names.index("Arac Tip Grubu")] = "Arac_Tip_Grubu"
names[names.index("Arac Tip")] = "Arac_Tip"
names[names.index("Model Yıl")] = "Model_Yıl"
names[names.index("Kasa Tipi")] = "Kasa_Tipi"
data.columns = names
data
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
```

```python
##burda yaptığımız işlem temizlenmiş datamızı içeren dataframe'inden aşağıdaki klonları silip
#varolan katagorik verileri yeni kolon ismi olarak belirledşkten sonra 0 ve 1  değerlerine atadık
dummy_Yakıt_Turu = pd.get_dummies(data["Yakıt_Turu"])
dummy_Vites = pd.get_dummies(data["Vites"])
dummy_Kimden = pd.get_dummies(data["Kimden"])
dummy_Durum = pd.get_dummies(data["Durum"])

data_new = data.copy()
data_new = pd.concat([data_new, dummy_Yakıt_Turu, dummy_Vites, dummy_Kimden, dummy_Durum], axis=1)
data_new.drop(["Yakıt_Turu", "Vites", "Kimden","Durum"], axis = 1, inplace=True)
#yeni vermiz;
data_new.head()
```

```python

#data_new=yeni datamız

#artı car_Data datafremi içerinde fiyat bilgisi yoktur.
car_data = data_new.drop(["Fiyat"],axis=1)

Fiyat = data_new["Fiyat"]

car_data["Fiyat"] = Fiyat

car_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
#verileri bilgisayarın anlayacağı şekle çeviriyoruz
le = LabelEncoder()
cols=['Marka', 'Arac_Tip_Grubu', 'Arac_Tip', 'CCM', 'Renk', 'Kasa_Tipi']

for i in cols:

    car_data[i] = le.fit_transform(car_data[i])#hem öğreneceksin hemde anlayacağı şekle çevirceksin.
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()
car_data.hist(ax = ax)
plt.show()
```

```python
car_data.head()#car_data=güncel datamız
```

```python
#veriler arasındaki ilişkileri inceliyoruz
plt.figure(figsize=(15,8))
sns.heatmap(car_data.corr(), annot = True)
plt.show()
```

```python
#verilerin eğitim ve test olarak bölünmesi
X_new= car_data.iloc[:,:-1]#en son kolon hariç diğerlerini al /fiyat haricinde diğer kolonlar
y_new= car_data.iloc[:,-1]#fiyat kolonu
X_new.head()
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X_new,y_new,test_size=.25, random_state=42)#X_new ve y_new değişkenlerini 
#yukarıda tanımladık
```

```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```


# Modelin Oluşturulması

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
#Sequential=modelimizi oluşturacağımız sınıf
#Dropout=bazı verilerin overfitting yapma işlemini yapmasını engellemek için bazı verilerin silinme işlemi yapılmıştır 
model.add(Dense(22,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(22,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(16,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(12,activation="relu"))

model.add(Dense(8,activation="relu"))

model.add(Dense(4,activation="relu"))

model.add(Dense(2,activation="relu"))

model.add(Dense(1))
#8 katman

model.compile(optimizer="adam", loss="mse")#Modelin tamamlanması
```

```python
#Erken Durdurma İşlemi;
earlyStopping = EarlyStopping(monitor="val_loss",mode="auto",verbose=1,patience=25)#earlystopping işlemi
#patience=25 -->25 defa tolerans göster demek
#Model Eğitim İşlemi
model.fit(X_train, y_train, validation_data=(X_test, y_test), verbose=1, epochs=45, callbacks=[earlyStopping])
#validation_data=doğrula
```

```python
#loss ve val_loss değerlerin incelenmesi
kayıpVeri = pd.DataFrame(model.history.history)
kayıpVeri.plot()
```

```python
#modelimizin sonucunun değerlendiriyoruz.
from sklearn.metrics import confusion_matrix#doğruluk ve yanılma payının hesaplanması
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score,precision_score,classification_report,roc_auc_score,roc_curve
```

```python
tahmin = model.predict(X_test)#predict=tahmin et
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mae_skor = mean_absolute_error(y_test, tahmin)
r2_skor = r2_score(y_test, tahmin)
hata_skor = np.sqrt(mean_squared_error(y_test, tahmin))

print("mae_skoru: ", mae_skor)
print("R2_skoru: ", r2_skor)
print("Hata Kare: ", hata_skor)
```
